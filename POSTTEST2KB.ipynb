```python
import matplotlib.pyplot as plt

datapie = data.target.value_counts()
plt.pie(datapie, labels=datapie.index, autopct="%1.1f%%")
plt.title("Presentase Penyakit Jantung")
plt.show()
```

**Distribusi pengidap penyakit jantung**
- 51.3% pasien memiliki penyakit jantung (target = 1)
- 48.7% pasien tidak memiliki penyakit jantung (target = 0)

```python
plt.hist([data[data['target'] == 0]['age'], data[data['target'] == 1]['age']], bins=10, alpha=0.7, label=['Tidak Sakit Jantung', 'Sakit Jantung'])
plt.title('Distribusi penyakit jantung berdasarkan umur', fontsize=12, fontweight='bold')
plt.legend()
plt.xlabel('Age')
plt.ylabel('Frekuensi')
plt.show()
```

**Kesimpulan**
Grafik menunjukkan bahwa penyakit jantung bisa menyerang berbagai usia, dengan konsentrasi tertinggi di usia 50-60 tahun. Kedua kelompok (sakit dan tidak sakit) memiliki distribusi yang overlap, menunjukkan bahwa usia saja tidak cukup untuk prediksi - perlu faktor lain.
